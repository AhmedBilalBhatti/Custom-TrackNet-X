```python
import os
import cv2
import torch
from ultralytics import YOLO  # YOLOv8 library

# Paths
dataset_path = r"D:\Computer Vision\FYP\TASK 1\env\Custom-TrackNet-X\DataSet\VeRi\image_train"
output_folder = r"D:\Computer Vision\FYP\TASK 1\env\Custom-TrackNet-X\DataSet\Refined\train"
os.makedirs(output_folder, exist_ok=True)

# Load YOLO model on GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = YOLO("yolov8n.pt")  # Replace with your YOLO model path
model.to(device)  # Send model to GPU if available

# Class ID for "car" (verify with your model's dataset; COCO default is 2)
CAR_CLASS_ID = 2

# Process images
no_detection_images = []
for img_name in os.listdir(dataset_path):
    img_path = os.path.join(dataset_path, img_name)

    # Read image
    image = cv2.imread(img_path)
    if image is None:
        print(f"Failed to load {img_name}")
        continue

    # Run YOLO detection (explicitly set the device for inference)
    results = model(image, device=device)

    # Filter for cars only
    cars = [box for box in results[0].boxes if int(box.cls) == CAR_CLASS_ID]

    # If no cars detected
    if len(cars) == 0:
        print(f"No car detected in: {img_name}")
        no_detection_images.append(img_name)
        continue

    # Crop and save each detected car
    for i, box in enumerate(cars):
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # Convert tensor to list if needed
        cropped = image[y1:y2, x1:x2]  # Crop bounding box

        # Save cropped image
        new_img_name = f"{os.path.splitext(img_name)[0]}_crop_{i}.jpg"
        save_path = os.path.join(output_folder, new_img_name)
        cv2.imwrite(save_path, cropped)

print("Processing complete!")
print("Images with no detections:", no_detection_images)
```
